## Exercise 1
Jane paid 12 dollars for 4 cups of coffee and 4 cups of tea. 3 cups of coffee cost as much as 2 cups of tea. What would be the total cost of 5 cups of coffee and 5 cups of tea?

### Solution

> Let $x$ be the unit price of coffee and $y$ be the unit price of tea

```python
# Your solution here 
# Answer: 5 cups of tea and 5 cups of coffee = 15 dollars
from sympy import symbols, Eq, solve

# Define the variables
# x = coffee
# y = tea
x, y = symbols('x y')

#Define the equations
# 1. 4x + 4y = 12  - 4 cups of tea and 4 cups of coffee = 12 dollars
# 2. 3x = 2y  - 3 cups of coffee cost the same as 2 cups of tea
eq1 = Eq(4*x + 4*y, 12)
eq2 = Eq(3*x, 2*y)

# Solve the equations
solution = solve((eq1, eq2), (x, y))
price_coffee = solution[x]
price_tea = solution[y]

# Total cost of 5 cups of tea and 5 cups of coffee
total_cost = 5 * price_coffee + 5 * price_tea
print('$', total_cost)

```

## Exercise 2

Jim has more money than Bob. If Jim gave Bob 20 dollars, they would have the same amount. If Bob gave Jim 22 dollars, however, Jim would then have twice as much as Bob. 

How much does each one actually have?

### Solution
> Let x be the amount of money that Jim has and y be the amount that Bob has 

```python
# Your solution here 
# Answer:
# y = 106 (Bob's amount)
# x = 146 (Jim's amount)

# Define the variables
# x = Jim's amount
# y = Bob's amount
x, y = symbols('x y')

# Define the equations
# 1. y + 20 = x - 20  - if Jim gives Bob 20 dollars, they will have the same amount
# 2. (x + 22) = (y - 22)*2) - if Bob gives Jim 22 dollars, Jim will have twice as much as Bob
eq1 = Eq(x-20, y+20)
eq2 = Eq(x+22, (y-22)*2)

# Solve the equations
solution = solve((eq1, eq2), (x, y))
jim_amount = solution[x]
bob_amount = solution[y]

print('Jim has $', jim_amount)
print('Bob has $', bob_amount)
```

## Exercise 3

Mia has 30 coins, consisting of quarters (25 cents) and dimes (10 cents), which totals to the amount 5.70 dollars.  
How many of each does she have?

### Solution

> Let x be the number of quarters and y be the number of dimes 

```python
# Your solution here 
# Answer:
# x = 18 quarters
# y = 12 dimes

# Define the variables
# x = number of quarters
# y = number of dimes
x, y = symbols('x y')

# Define the equations
# 1. x + y = 30 - total number of coins is 30
# 2. 0.25x + 0.10y = 5.70 - total value of coins is 5.70
eq1 = Eq(x+y, 30)
eq2 = Eq(0.25*x + 0.10*y, 5.70)

# Solve the equations
solution = solve((eq1, eq2), (x, y))
quarters = solution[x]
dimes = solution[y]

print('Number of quarters:', quarters)
print('Number of dimes:', dimes)
```
